```python
myconnection = pymysql.connect(host = '127.0.0.1',user='root',passwd='Ishu@0208',database = "NASA_NEO_PROJECT")
```

```python
cur = myconnection.cursor()
```

```python
cur.execute("create table asteroids(id int,name varchar(50),absolute_magnitude_h float,estimated_diameter_min_km float,estimated_diameter_max_km float,is_potentially_hazardous_asteroid boolean)")
```

```python
cur.execute("create table close_approach(neo_reference_id int,close_approach_date date,relative_velocity_kmph float,astronomical float,miss_distance_km float,miss_distance_lunar float,orbiting_body varchar(20))")
```

```python
#first table: asteroids
insert_query = """
    insert ignore into asteroids (
        id,
        name,
        absolute_magnitude_h,
        estimated_diameter_min_km,
        estimated_diameter_max_km,
        is_potentially_hazardous_asteroid
    ) VALUES (%s, %s, %s, %s, %s, %s);
"""

#second table: close_approach
insert_query1 = """
    insert ignore into close_approach (
        neo_reference_id,
        close_approach_date,
        relative_velocity_kmph,
        astronomical,
        miss_distance_km,
        miss_distance_lunar,
        orbiting_body
    ) VALUES (%s, %s, %s, %s, %s, %s, %s);
"""

#Loop through asteroids_data and insert into both tables
for ast in asteroids_data:
    #first insert into asteroids table
    values = (
        ast['id'],
        ast['name'],
        ast['absolute_magnitude_h'],
        ast['estimated_diameter_min_km'],
        ast['estimated_diameter_max_km'],
        ast['is_potentially_hazardous_asteroid']
    )
    try:
        cur.execute(insert_query, values)
    except Exception as e:
        print(f"Insert failed (asteroid) for {ast['id']}: {e}")

#second insert into close_approach table
    values1 = (
        ast['neo_reference_id'],
        ast['close_approach_date'],
        ast['relative_velocity_kmph'],
        ast['astronomical'],
        ast['miss_distance_km'],
        ast['miss_distance_lunar'],
        ast['orbiting_body']
    )
    try:
        cur.execute(insert_query1, values1)
    except Exception as e:
        print(f"Insert failed (close_approach) for {ast['neo_reference_id']}: {e}")

#Final part
myconnection.commit()

print("All data inserted successfully.")
```

```python
#1 Count how many times each asteroid has approached Earth
```

```python
pd.read_sql_query("select neo_reference_id, count(*) as approach_count from close_approach where orbiting_body='Earth' group by neo_reference_id",myconnection)
```

```python
#2 Average velocity of each asteroid over multiple approaches
```

```python
pd.read_sql_query("select neo_reference_id, avg(relative_velocity_kmph) as average_velocity from close_approach group by neo_reference_id",myconnection)
```

```python
#3 List top 10 fastest asteroids
```

```python
pd.read_sql_query("select distinct a.id,a.name,c.relative_velocity_kmph from close_approach as c join asteroids as a on c.neo_reference_id = a.id where c.orbiting_body = 'Earth' order by c.relative_velocity_kmph desc limit 10",myconnection)
```

```python
#4 Find potentially hazardous asteroids that have approached Earth more than 3 times
```

```python
pd.read_sql_query("select a.id,a.name from asteroids as a join close_approach as c on a.id = c.neo_reference_id where a.is_potentially_hazardous_asteroid = true and c.orbiting_body = 'Earth' group by a.id, a.name having COUNT(c.neo_reference_id) > 3",myconnection)
```
